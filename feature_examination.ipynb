```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data/train.csv")
df_test = pd.read_csv("Data/test.csv")
```

```python
df.groupby(['Age'])['Survived'].mean()
```

```python
titanic_df = df.dropna(subset=['Age', 'Survived'])
dizi = np.arange(0, 101, 10)


titanic_df['age_category'] = pd.cut(titanic_df['Age'], bins=dizi)

# Her yaş kategorisi için hayatta kalma oranını hesaplayalım
survival_rates = titanic_df.groupby('age_category')['Survived'].mean()

# Grafiği çizelim
plt.figure(figsize=(15, 6))
sns.barplot(x=survival_rates.index, y=survival_rates.values)
plt.title('Hayatta Kalma Oranı Yaş Kategorisine Göre')
plt.xlabel('Yaş Kategorisi')
plt.ylabel('Hayatta Kalma Oranı')
plt.show()
```
